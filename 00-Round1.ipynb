```python
# -1: ambiguous
df['SuperBin'] = -1

df.head()
```

```python
def determine_superbin(row):
    low = row[('Binned', 'low3bins')]
    medium = row[('Binned', 'medium3bins')]
    high = row[('Binned', 'high3bins')]
    
    if low == 0 and medium == 0 and high == 0:
        return 0
    elif low == 1 and medium == 0 and high == 0:
        return 1
    elif (low == 2 and medium == 0 and high == 0) or (low == 1 and medium == 1 and high == 0):
        return 2
    elif low == 2 and medium == 1 and high == 0:
        return 3
    elif (low == 2 and medium == 2 and high == 0) or (low == 2 and medium == 1 and high == 1):
        return 4
    elif low == 2 and medium == 2 and high == 1:
        return 5
    elif low == 2 and medium == 2 and high == 2:
        return 6
    else:
        return -1  # Ambiguous state
```

```python
df['SuperBin'] = df.apply(determine_superbin, axis=1)
```

```python
# 0: all 0
df.loc[(df[('Binned', 'low3bins')] == 0) &
        (df[('Binned', 'medium3bins')] == 0) &
        (df[('Binned', 'high3bins')] == 0), ('SuperBin', '')] = 0
```
